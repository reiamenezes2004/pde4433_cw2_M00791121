```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
char_folder = "segmented_plates"
output_csv = "character_labels.csv"

labeled_data = []
char_images = sorted([f for f in os.listdir(char_folder) if f.endswith(".png")])

print("Total characters to label:", len(char_images))

for img_file in char_images:
    img_path = os.path.join(char_folder, img_file)
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

    # Display the character image
    plt.imshow(img, cmap='gray')
    plt.title(f"Label for: {img_file}")
    plt.axis('off')
    plt.show()

    label = input(f"Enter label for {img_file}: ").strip().upper()
    labeled_data.append([img_file, label])

# Save to CSV
df = pd.DataFrame(labeled_data, columns=["filename", "label"])
df.to_csv(output_csv, index=False)
print("All labels saved to", output_csv)
```
